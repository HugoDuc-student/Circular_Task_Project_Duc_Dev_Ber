```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- CONFIGURATION ---
FILE = "data.csv"
marker = "marker.csv"

# --- lecture functions

def extract_header_params(file_path):
    params = {}
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            header_line = f.readline().strip()
            parts = header_line.split(';')
            for part in parts:
                if ' ' in part:
                    try:
                        key, val = part.strip().rsplit(' ', 1)
                        try: params[key] = float(val)
                        except ValueError: params[key] = val
                    except: pass
    except Exception as e:
        print(f"Impossible to read all parameters ({e})")
    return params
# --- print head line information ---
with open(FILE, 'r', encoding='utf-8') as f:
    firstLine = f.readline().strip()

if firstLine : 
    data = firstLine.split(';')
    print("Head line information in data file:")
    for d in data : 
        print(" " + d.strip())

#print marker infos from line 4 to 26 and column 2 to 9
marker_file = marker
with open(marker_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()[3:26]
    print("\nMarker information in the file.csv:")
    for line in lines:
        data = line.strip().split(';')[1:9]
        print(" ".join(data))


def load_data_robust(file_path):
    """Load data ignoring text lines and detecting the separator."""
    valid_data = []
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
        
        print(f"Reading files {file_path} ({len(lines)} lines)...")

        # Separator detection
        sample = lines[10:110] if len(lines) > 110 else lines
        delimiter = ';' if sum(l.count(';') for l in sample) >= sum(l.count(',') for l in sample) else ','
        print(f"-> Detected separator: '{delimiter}'")
        
        for line in lines:
            cols = line.strip().split(delimiter)
            # Cleaning and checking if the first column is a digit
            first_col = cols[0].replace('.', '', 1).replace('-', '', 1)
            if len(cols) >= 4 and first_col.isdigit():
                try:
                    row = [float(c) if c.strip() else np.nan for c in cols[:4]]
                    valid_data.append(row)
                except ValueError:
                    continue 
                    
    except FileNotFoundError:
        print(f"ERROR: File {file_path} not found.")
        return None

    data = np.array(valid_data)
    if data.size == 0:
        print("ERROR: No valid numeric data found.")
        return None
        
    print(f"-> Data loaded successfully ({len(data)} points).")
    return data[:, 0], data[:, 1], data[:, 2], data[:, 3]


def sanitize_timestamps(ts):
    """Repair Corrupted Timestamps (non-increasing)."""
    diffs = np.diff(ts)
    if np.sum(diffs <= 0) > 0:
        corrected_ts = [ts[0]]
        for d in diffs:
            gap = abs(d)
            if gap < 0.0001 or gap > 3600000: gap = 8.0 
            corrected_ts.append(corrected_ts[-1] + gap)
        return np.array(corrected_ts)
    return ts

def find_records_clinical(ts, cycle_duration=20.0):
    if len(ts) < 10: return [], 1.0, ts
    
    ts = sanitize_timestamps(ts)
    
    # Détection unité de temps
    raw_diff_median = np.median(np.diff(ts))
    if raw_diff_median < 1.0:
        time_factor = 1.0
        unit = "Seconds"
    else:
        time_factor = 1000.0
        unit = "Milliseconds"
        
    print(f"-> Time units : {unit} (Factor: {time_factor})")
    
    ts_sec = ts / time_factor
    diffs = np.diff(ts_sec)
    
    # Clinical threshold (> 5s)
    CLINICAL_THRESHOLD = max(5.0, cycle_duration * 0.25)
    print(f"-> Cut-off threshold (End of trial) : {CLINICAL_THRESHOLD:.1f} seconds")
    
    # Logic i+1 for boundaries
    breaks = np.where(diffs > CLINICAL_THRESHOLD)[0] + 1
    boundaries = [0] + list(breaks) + [len(ts)]
    
    records = []
    for i in range(len(boundaries)-1):
        start = boundaries[i]
        end = boundaries[i+1]
        
        duration = ts_sec[end-1] - ts_sec[start]
        if duration > 1.0: 
            records.append((start, end))
        else:
            if i < 3: print(f"   [Info] Segment ignoré (bruit de {duration:.3f}s)")
            
    if len(records) == 0 and len(ts) > 200:
        print("-> Aucun break détecté. Analyse globale.")
        records.append((0, len(ts)))
            
    return records, time_factor, ts

# Print duration of each detected record (seconds)
res = load_data_robust(FILE)
if res is None:
    print('No data loaded')
else:
    ts_raw, mx, my, mit = res
    records, time_factor, ts = find_records_clinical(ts_raw, cycle_duration=20.0)
    ts_sec = ts / time_factor
    if not records:
        print('No records detected')
    else:
        for i, (start, end) in enumerate(records, start=1):
            # duration using last timestamp of segment minus first
            duration = ts_sec[end-1] - ts_sec[start]
            print(f'Record {i}: start={start}, end={end}, duration={duration:.2f} s')


# --- 2. PRECISE CLEANING FUNCTION (TRIMMING) --- 

def trim_to_active_movement(t, x, y, mit):
    if len(t) < 10: return t, x, y, mit

    # 1. Velocity
    dt = np.diff(t)
    dt = np.where(dt == 0, 1e-9, dt)
    vel = np.sqrt(np.diff(x)*2 + np.diff(y)*2) / dt
    
    # 2. Smoothing (Moving average over 10 points)
    window = 10
    if len(vel) > window:
        vel_smooth = np.convolve(vel, np.ones(window)/window, mode='same')
    else:
        vel_smooth = vel

    # 3. Movement threshold (> 15 px/s)
    is_moving = vel_smooth > 15.0
    
    if np.sum(is_moving) == 0:
        return t, x, y, mit
        
    # 4. Find boundaries
    indices = np.where(is_moving)[0]
    first_idx = indices[0]
    last_idx = min(indices[-1] + 5, len(t)) # End margin
    
    return t[first_idx:last_idx], x[first_idx:last_idx], y[first_idx:last_idx], mit[first_idx:last_idx]

# --- 3. CALCULATION FUNCTIONS ---
def compute_metrics(t, x, y, target):
    radius = np.sqrt(x*2 + y*2)
    dt = np.diff(t)
    dt = np.where(dt == 0, 1e-9, dt)
    vx = np.diff(x) / dt
    vy = np.diff(y) / dt
    velocity = np.sqrt(vx*2 + vy*2)
    velocity = np.append(velocity, 0)
    success_rate = (np.sum(target == 1) / len(target)) * 100
    return radius, velocity, success_rate

# --- 4. MAIN ---

def main():
    print(f"--- Clinical analysis : {FILE} ---\n")
    
    # A. Parameters
    params = extract_header_params(FILE)
    CX = params.get('centerX', 951.0)
    CY = params.get('centerY', 486.0)
    R_EXT = params.get('externalRadius', 406.0) - params.get('cursorRadius', 16)
    R_INT = params.get('internalRadius', 326.0) + params.get('cursorRadius', 16)
    CYCLE_DUR = params.get('cycleDuration', 20.0)
    
    print(f"Paramètres : Centre({CX}, {CY}) | Rayons({R_INT:.0f}-{R_EXT:.0f})")

    # B. Loading data
    res = load_data_robust(FILE)
    if res is None: return
    ts_raw, mx, my, mit = res
    
    xc = mx - CX
    yc = my - CY 
    
    # C. Segmentation
    records, time_factor, ts = find_records_clinical(ts_raw, cycle_duration=CYCLE_DUR)
    
    print(f"Records found: {len(records)}\n")
    if not records:
        print("ERROR: No usable data.")
        return

    # D. Generating records
    for i, (start, end) in enumerate(records):
        
        # 1. Raw extraction
        ts_sec = ts / time_factor
        t_seg_raw = ts_sec[start:end]
        x_seg_raw = xc[start:end]
        y_seg_raw = yc[start:end]
        mit_seg_raw = mit[start:end]
        
        # 2. PRECISE CLEANING (TRIMMING)
        t_seg, x_seg, y_seg, mit_seg = trim_to_active_movement(t_seg_raw, x_seg_raw, y_seg_raw, mit_seg_raw)
        
        # Realignment
        t_seg = t_seg - t_seg[0]
        duration = t_seg[-1]
        
        # Validation
        status = "✅ Valid" if (duration > 19.0 and duration < 26.0) else "⚠ Atypical Duration"
        print(f"Trial {i+1} : Active Duration = {duration:.2f}s -> {status}")
        
        # Metrics
        radii, velocity, success = compute_metrics(t_seg, x_seg, y_seg, mit_seg)
        y_seg_plot = -y_seg # Inversion for screen display
        
        # --- LAYOUT ---
        fig = plt.figure(figsize=(16, 10))
        gs = fig.add_gridspec(3, 2)
        fig.suptitle(f" Trial n°{i+1} (Active Duration: {duration:.1f}s)", fontsize=16, fontweight='bold')
        
        # LEFT: TRAJECTORY
        ax_traj = fig.add_subplot(gs[:, 0])
        ax_traj.set_aspect('equal')
        
        ax_traj.add_patch(patches.Circle((0, 0), R_EXT, color='yellow', alpha=0.15))
        ax_traj.add_patch(patches.Circle((0, 0), R_INT, color='white'))
        ax_traj.add_patch(patches.Circle((0, 0), R_EXT, color='red', fill=False, lw=1))
        ax_traj.add_patch(patches.Circle((0, 0), R_INT, color='red', fill=False, lw=1))
        
        mask_in = mit_seg == 1
        ax_traj.scatter(x_seg[mask_in], y_seg_plot[mask_in], c='green', s=10, alpha=0.5, label='Succès')
        ax_traj.scatter(x_seg[~mask_in], y_seg_plot[~mask_in], c='red', s=15, alpha=0.7, label='Erreur')
        
        # FIXED ZOOM REQUESTED
        ax_traj.set_xlim(-750, 750)
        ax_traj.set_ylim(-750, 750)
        
        ax_traj.set_title(f"Spatial trajectory (Precision: {success:.1f}%)")
        ax_traj.set_xlabel("X (pixels)")
        ax_traj.set_ylabel("Y (pixels)")
        ax_traj.legend(loc='upper right')
        ax_traj.grid(True, linestyle=':', alpha=0.6)

        # curve 1 : X OVER TIME
        ax_x = fig.add_subplot(gs[0, 1])
        ax_x.plot(t_seg, x_seg, color='tab:blue', lw=1.5)
        ax_x.axhline(0, color='black', lw=0.5)
        ax_x.set_title("X over Time")
        ax_x.set_ylabel("Pos X (px)")
        ax_x.grid(True, alpha=0.3)

        # curve 2 : Y OVER TIME
        ax_y = fig.add_subplot(gs[1, 1], sharex=ax_x)
        ax_y.plot(t_seg, y_seg, color='tab:orange', lw=1.5) 
        ax_y.axhline(0, color='black', lw=0.5)
        ax_y.set_title("Y over Time")
        ax_y.set_ylabel("Pos Y (px)")
        ax_y.grid(True, alpha=0.3)

        # curve 3 : TARGET STATUS
        ax_mit = fig.add_subplot(gs[2, 1], sharex=ax_x)
        ax_mit.fill_between(t_seg, 0, mit_seg, color='green', alpha=0.3, step='mid')
        ax_mit.plot(t_seg, mit_seg, color='darkgreen', lw=1.5, drawstyle='steps-mid')
        
        ax_mit.set_yticks([0, 1])
        ax_mit.set_yticklabels(['Hors', 'Dedans'])
        ax_mit.set_ylim(-0.1, 1.1)
        ax_mit.set_title("Mouse In Target (Status)")
        ax_mit.set_xlabel("Time (s)")
        ax_mit.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

main()
```

```python
# Final overall plot: single circle overlay + combined Position vs Time (20s windows) with 20s pause between records
MAX_SEG_SEC = 20.0
GAP_SECONDS = 20.0  # pause between records
params = extract_header_params(FILE)
CX = params.get('centerX', 951.0)
CY = params.get('centerY', 486.0)
R_EXT = params.get('externalRadius', 406.0) - params.get('cursorRadius', 16)
R_INT = params.get('internalRadius', 326.0) + params.get('cursorRadius', 16)

res = load_data_robust(FILE)
if res is None:
    print('No data to plot')
else:
    ts_raw, mx, my, mit = res
    xc = mx - CX
    yc = my - CY

    records, time_factor, ts = find_records_clinical(ts_raw, cycle_duration=MAX_SEG_SEC)
    if not records:
        print('No records detected')
    else:
        # Build truncated rounds (T_rel, x, y, mit) capped to MAX_SEG_SEC
        rounds = []
        ts_sec_all = ts / time_factor
        for (start, end) in records:
            t_seg = ts_sec_all[start:end]
            if len(t_seg) == 0:
                continue
            t_rel = t_seg - t_seg[0]
            end_idx = np.searchsorted(t_rel, MAX_SEG_SEC, side='right')
            rounds.append((t_rel[:end_idx], xc[start:end][:end_idx], yc[start:end][:end_idx], mit[start:end][:end_idx]))

        n = len(rounds)
        import matplotlib.cm as cm
        cmap = cm.get_cmap('tab20', max(1, n))

        # compute offsets with gaps between records
        offsets = np.arange(n) * (MAX_SEG_SEC + GAP_SECONDS)
        total_time = n * MAX_SEG_SEC + max(0, n-1) * GAP_SECONDS

        # helper: plot contiguous segments colored by mask (True -> green inside, False -> red outside)
        def plot_segments(ax, x, y, mask, lw=1.5):
            if len(x) == 0:
                return
            i = 0
            L = len(x)
            while i < L:
                cur = bool(mask[i])
                j = i + 1
                while j < L and bool(mask[j]) == cur:
                    j += 1
                if j - i >= 2:
                    ax.plot(x[i:j], -y[i:j], color=('green' if cur else 'red'), lw=lw, alpha=0.9)
                else:
                    ax.scatter(x[i], -y[i], color=('green' if cur else 'red'), s=18, alpha=0.9)
                i = j

        # Figure with left circle and right time-series
        fig = plt.figure(figsize=(18, 9))
        gs = fig.add_gridspec(1, 2, width_ratios=[1, 1.4], wspace=0.3)
        ax_circle = fig.add_subplot(gs[0, 0])
        ax_time = fig.add_subplot(gs[0, 1])

        # Draw target rings on circle axis
        ax_circle.add_patch(patches.Circle((0, 0), R_EXT, edgecolor='yellow', facecolor='none', lw=2))
        ax_circle.add_patch(patches.Circle((0, 0), R_INT, edgecolor='yellow', facecolor='none', lw=2))
        # 2. Target (Yellow Background)
        
        ax_circle.set_aspect('equal')
        ax_circle.set_xlim(-750, 750)
        ax_circle.set_ylim(-750, 750)
        ax_circle.set_title('All records overlay on circle (20s max each)')
        ax_circle.set_xlabel('X (pixels)')
        ax_circle.set_ylabel('Y (pixels)')

        # Time axis: will cover 0..total_time including gaps
        ax_time.set_xlim(0-10, total_time+10)
        ax_time.set_xlabel('Time (s)')
        ax_time.set_ylabel('Centered position (px)')
        ax_time.set_title('Position over time — 20s windows with 20s pause between records')

        color_x = 'tab:blue'
        color_y = 'tab:orange'

        for i, (T, X, Y, trg) in enumerate(rounds):
            if len(T) == 0:
                continue
            # LEFT: overlay trajectories, plot contiguous inside/outside segments in green/red
            mask_in = (trg == 1)
            plot_segments(ax_circle, X, Y, mask_in, lw=1.8)

            # RIGHT: place each record into its fixed window with gap
            offset = offsets[i]
            T_plot = T + offset
            ax_time.plot(T_plot, X, color=color_x, lw=1.2, alpha=0.9, label='X' if i == 0 else '_nolegend_')
            ax_time.plot(T_plot, -Y, color=color_y, lw=1.2, alpha=0.9, label='Y' if i == 0 else '_nolegend_')

            # vertical separators for this window (light)
            ax_time.axvline(offset, color='gray', ls='--', lw=0.6, alpha=0.7)
            ax_time.axvline(offset + MAX_SEG_SEC, color='gray', ls='--', lw=0.6, alpha=0.7)

            # solid colored bars: black at start, red at end (visible, thicker, top zorder)
            ax_time.axvline(offset, color='black', ls='--', lw=1, alpha=1.0, zorder=6)
            ax_time.axvline(offset + MAX_SEG_SEC, color='red', ls='--', lw=1, alpha=1.0, zorder=6)

        # finalize legends and layout
        from matplotlib.lines import Line2D
        legend_elements = [
            Line2D([0], [0], color='green', lw=3, label='inside target'),
            Line2D([0], [0], color='red', lw=3, label='outside target'),
            patches.Patch(facecolor='yellow', alpha=0.4, label = 'Target')
        ]
        ax_circle.legend(handles=legend_elements, loc='upper right', fontsize='small')
        ax_time.legend(loc='upper right', fontsize='small')
        ax_time.grid(True, linestyle=':', alpha=0.4)
        
        plt.suptitle('Overall: Circle (left) and Position/Time (right) — each record mapped to 20s window with 20s pause between records')
        plt.tight_layout(rect=[0,0,1,0.96])
        plt.show()
```
